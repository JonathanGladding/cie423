```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
delta = np.linspace(0, 10)
alpha1 = 0
alpha2 = 0.2

def P(alpha):
    return (
    1
    / 6
    * (
        (6 * (36 + delta**2 - 6 * np.sqrt(2) * delta) ** (-1 / 2) - 1)
        * 1200000
        * (3 * np.sqrt(2) - delta)
        + 600000 * alpha * delta
    )
)
```

```python
f, (ax1, ax2) = plt.subplots(1, 2, sharey=True)
ax1.plot(delta, P(alpha1))
ax1.set_title('P-Delta Effect: No Spring')
ax2.plot(delta, P(alpha2))
ax2.set_title('P-Delta Effect: With Spring')
```
